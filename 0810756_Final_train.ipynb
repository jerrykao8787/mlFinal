## Load data


```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
from sklearn.model_selection import train_test_split
```

```python
train_df = pd.read_csv('train.csv')
test_df = pd.read_csv('test.csv')

print(train_df.shape)
print(test_df.shape)
```

```python
traindata = train_df.append(test_df).iloc[:,2:]
traindata.drop('attribute_1', inplace=True, axis=1)
traindata['attribute_0']=pd.get_dummies(traindata['attribute_0'], columns=['attribute_0'], drop_first=True)

traindataNP = traindata.to_numpy()
```

```python
x_train = traindataNP[:26570,:-1] 
y_train = traindataNP[:26570,-1]

print(x_train.shape)#26570 22
print(y_train.shape)#26570
```

```python
imr = SimpleImputer(missing_values=np.nan, strategy='mean', copy=False)
imr = imr.fit_transform(x_train)
```

```python
x_train, x_val, y_train, y_val = train_test_split(x_train, y_train, test_size=0.1, random_state=42)
```

```python
import torch

train_features = torch.Tensor(x_train)
train_targets = torch.Tensor(y_train.reshape(-1,1))
val_features = torch.Tensor(x_val)
val_targets = torch.Tensor(y_val.reshape(-1,1))
```

```python
from torch.utils.data import TensorDataset, DataLoader

batch_size = 30000

train = TensorDataset(train_features, train_targets)
val = TensorDataset(val_features, val_targets)

train_loader = DataLoader(train, batch_size=batch_size, shuffle=True, drop_last=False, num_workers=2, pin_memory=True)
val_loader = DataLoader(val, batch_size=batch_size, shuffle=False, drop_last=False, num_workers=2, pin_memory=True)
```

```python
import torch
import torch.nn as nn

class Classifier(nn.Module):
    def __init__(self):
        super(Classifier, self).__init__()
        self.layer1A = nn.Sequential(nn.Linear(22, 22),nn.ReLU(), nn.BatchNorm1d(22), nn.Dropout(p=0.2) ,nn.Linear(22, 20), nn.ReLU())
        self.dense2 = nn.Sequential(nn.Linear(20, 19), nn.ReLU())
        self.dense3 = nn.Sequential(nn.Linear(19, 18), nn.ReLU())
        self.dense4 = nn.Sequential(nn.Linear(18, 17), nn.ReLU())
        self.dense5 = nn.Sequential(nn.Linear(17, 16), nn.ReLU(), nn.Linear(16, 15), nn.ReLU(), nn.BatchNorm1d(15))
        self.dense6 = nn.Sequential(nn.Linear(17, 15), nn.ReLU(), nn.Linear(15, 14), nn.ReLU(), nn.BatchNorm1d(14))
        self.dense7 = nn.Sequential(nn.Linear(18, 16), nn.ReLU())
        self.dense8 = nn.Sequential(nn.Linear(16, 15), nn.ReLU(), nn.Linear(15, 14), nn.ReLU(), nn.BatchNorm1d(14))
        self.dense9 = nn.Sequential(nn.Linear(16, 14), nn.ReLU(), nn.Linear(14, 13), nn.ReLU(), nn.BatchNorm1d(13))
        self.dense10 = nn.Sequential(nn.Linear(19, 17), nn.ReLU())
        self.dense11 = nn.Sequential(nn.Linear(17, 16), nn.ReLU())
        self.dense12 = nn.Sequential(nn.Linear(16, 15), nn.ReLU(), nn.Linear(15, 14), nn.ReLU(), nn.BatchNorm1d(14))
        self.dense13 = nn.Sequential(nn.Linear(16, 14), nn.ReLU(), nn.Linear(14, 13), nn.ReLU(), nn.BatchNorm1d(13))
        self.dense14 = nn.Sequential(nn.Linear(17, 15), nn.ReLU())
        self.dense15 = nn.Sequential(nn.Linear(15, 14), nn.ReLU(), nn.Linear(14, 13), nn.ReLU(), nn.BatchNorm1d(13))
        self.dense16 = nn.Sequential(nn.Linear(15, 13), nn.ReLU(), nn.Linear(13, 12), nn.ReLU(), nn.BatchNorm1d(12))
        self.dense17 = nn.Sequential(nn.Linear(20, 18), nn.ReLU())
        self.dense18 = nn.Sequential(nn.Linear(18, 17), nn.ReLU())
        self.dense19 = nn.Sequential(nn.Linear(17, 16), nn.ReLU())
        self.dense20 = nn.Sequential(nn.Linear(16, 15), nn.ReLU(), nn.Linear(15, 14), nn.ReLU(), nn.BatchNorm1d(14))
        self.dense21 = nn.Sequential(nn.Linear(16, 14), nn.ReLU(), nn.Linear(14, 13), nn.ReLU(), nn.BatchNorm1d(13))
        self.dense22 = nn.Sequential(nn.Linear(17, 15), nn.ReLU())
        self.dense23 = nn.Sequential(nn.Linear(15, 14), nn.ReLU(), nn.Linear(14, 13), nn.ReLU(), nn.BatchNorm1d(13))
        self.dense24 = nn.Sequential(nn.Linear(15, 13), nn.ReLU(), nn.Linear(13, 12), nn.ReLU(), nn.BatchNorm1d(12))
        self.dense25 = nn.Sequential(nn.Linear(18, 16), nn.ReLU())
        self.dense26 = nn.Sequential(nn.Linear(16, 15), nn.ReLU())
        self.dense27 = nn.Sequential(nn.Linear(15, 14), nn.ReLU(), nn.Linear(14, 13), nn.ReLU(), nn.BatchNorm1d(13))
        self.dense28 = nn.Sequential(nn.Linear(15, 13), nn.ReLU(), nn.Linear(13, 12), nn.ReLU(), nn.BatchNorm1d(12))
        self.dense29 = nn.Sequential(nn.Linear(16, 14), nn.ReLU())
        self.dense30 = nn.Sequential(nn.Linear(14, 13), nn.ReLU(), nn.Linear(13, 12), nn.ReLU(), nn.BatchNorm1d(12))
        self.dense31 = nn.Sequential(nn.Linear(14, 12), nn.ReLU(), nn.Linear(12, 11), nn.ReLU(), nn.BatchNorm1d(11))
        self.dense32 = nn.Sequential(nn.Linear(208, 20), nn.ReLU(), nn.Linear(20, 1), nn.Sigmoid())

    def forward(self, x):
        x = self.layer1A(x)
        d2out = self.dense2(x)
        d10out = self.dense10(d2out)
        d11out = self.dense11(d10out)
        d12out = self.dense12(d11out)
        d13out = self.dense13(d11out)
        d14out = self.dense14(d10out)
        d15out = self.dense15(d14out)
        d16out = self.dense16(d14out)
        d3out = self.dense3(d2out)
        d4out = self.dense4(d3out)
        d7out = self.dense7(d3out)
        d5out = self.dense5(d4out)
        d6out = self.dense6(d4out)
        d8out = self.dense8(d7out)
        d9out = self.dense9(d7out)
        d17out = self.dense17(x)
        d18out = self.dense18(d17out)
        d19out = self.dense19(d18out)
        d22out = self.dense22(d18out)
        d20out = self.dense20(d19out)
        d21out = self.dense21(d19out)
        d23out = self.dense23(d22out)
        d24out = self.dense24(d22out)
        d25out = self.dense25(d17out)
        d26out = self.dense26(d25out)
        d29out = self.dense29(d25out)
        d27out = self.dense27(d26out)
        d28out = self.dense28(d26out)
        d30out = self.dense30(d29out)
        d31out = self.dense31(d29out)
        catout = torch.cat((d5out, d6out, d8out, d9out, d12out, d13out, d15out, d16out, d20out, d21out, d23out, d24out, d27out, d28out, d30out, d31out), dim=1)
        catout = self.dense32(catout)
        
        return catout
```

```python
device = 'cuda' if torch.cuda.is_available() else 'cpu'
model_path = './model.ckpt'

# create model, define a loss function, and optimizer
model = Classifier().to(device)
criterion = nn.BCELoss()
```

```python
learning_rate = 0.001
optimizer = torch.optim.Adam(model.parameters(), lr=learning_rate, weight_decay=0)
scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(optimizer, mode='min', factor=0.5, patience=8, min_lr=1e-12)
num_epoch = 300
```

```python
# start training

best_loss = 10000.0
for epoch in range(num_epoch):
    train_acc = 0.0
    train_loss = 0.0
    val_acc = 0.0
    val_loss = 0.0

    # training
    model.train() # set the model to training mode
    for i, data in enumerate(train_loader):
        inputs, labels = data
        inputs, labels = inputs.to(device), labels.to(device)
        optimizer.zero_grad() 
        outputs = model(inputs) 
        batch_loss = criterion(outputs, labels)
        batch_loss.backward() 
        optimizer.step() 
        train_loss += batch_loss.item()

    # validation
    if len(val) > 0:
        model.eval() # set the model to evaluation mode
        with torch.no_grad():
            for i, data in enumerate(val_loader):
                inputs, labels = data
                inputs, labels = inputs.to(device), labels.to(device)
                outputs = model(inputs)
                batch_loss = criterion(outputs, labels)
                val_loss += batch_loss.item()

            print('[{:03d}/{:03d}] Train Loss: {:3.6f} | Val loss: {:3.6f}'.format(
                epoch + 1, num_epoch, train_loss/len(train_loader), val_loss/len(val_loader)
            ))
            scheduler.step(val_loss)
            # if the model improves, save a checkpoint at this epoch
            if val_loss < best_loss:
                best_loss = val_loss
                torch.save(model.state_dict(), model_path)
                print('saving model with loss {:.3f}'.format(best_loss/len(val_loader)))
```
